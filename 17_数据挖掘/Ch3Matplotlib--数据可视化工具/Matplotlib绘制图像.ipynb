## 绘图步骤
画图流程大致分为如下步骤：
- 使用figure()函数创建画布，决定是否创建子图
- 使用plot()函数绘制图形
- 设置绘图对象的各种属性

### 创建画布

```python
import matplotlib.pyplot as plt  

# 创建一个新的图形，编号为 1，大小为 8x6 英寸  
plt.figure(num=1, figsize=(8, 6), dpi=100, facecolor='white', edgecolor='blue')   

# 在 figure 1 中创建一个子图  
plt.plot([1, 2, 3, 4, 5], [2, 4, 6, 8, 10],color='red')  

# 添加标题和标签  
plt.title("Simple Line Plot")  
plt.xlabel("X-axis")  
plt.ylabel("Y-axis")  

# 显示图形  
plt.show()
```

### 子图操作

```python
import numpy as np  
import matplotlib.pyplot as plt  

x = np.linspace(-np.pi, np.pi, 200)  
cos_y = np.cos(x) / 2  
sin_y = np.sin(x)  
f = cos_y + sin_y  
z = cos_y - sin_y  

# 创建画布
plt.figure(num=1, figsize=(12, 8), dpi=400, facecolor='white') 

# 总标题  
plt.suptitle("Trigonometric Functions and Their Combinations", fontsize=16, color='darkred')  

# 子图1  
plt.subplot(221)  
plt.plot(x, cos_y, linestyle='--', linewidth=1, color='g')  
plt.xlabel("x")  
plt.ylabel("cos(x)/2")  
plt.title("Cosine Function (Scaled)")  

# 子图2  
plt.subplot(222)  
plt.plot(x, sin_y, linestyle=':', linewidth=1, color='blue')  
plt.xlabel("x")  
plt.ylabel("sin(x)")  
plt.title("Sine Function")  

# 子图3  
plt.subplot(223)  
plt.plot(x, f, linestyle='-', linewidth=1, color='black')  
plt.xlabel("x")  
plt.ylabel("cos(x)/2 + sin(x)")  
plt.title("Sum of Functions")  

# 子图4  
plt.subplot(224)  
plt.plot(x, z, linestyle='-.', linewidth=1, color='pink')  
plt.xlabel("x")  
plt.ylabel("cos(x)/2 - sin(x)")  
plt.title("Difference of Functions")  

plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # 调整子图布局，为suptitle腾出空间  
plt.show()
```

```python
import matplotlib.pyplot as plt  
import numpy as np  

# 创建一些示例数据  
np.random.seed(42)  # 为了可重复性  
data = np.random.randn(1000)  
labels = ['A', 'B', 'C', 'D']  
sizes = [15, 30, 45, 10]  
x = np.random.rand(50)  
y = np.random.rand(50)  
categories = ['Category 1', 'Category 2', 'Category 3']  
values = [25, 40, 35]  

# 创建图形  
plt.figure(figsize=(12, 8), dpi=400)  
plt.suptitle("Combined Plots", fontsize=16)  

# 子图 1: 直方图  
plt.subplot(2, 2, 1)  # 2行2列的第一个子图  
plt.hist(data, bins=30, color='skyblue', edgecolor='black')  
plt.title('Histogram')  
plt.xlabel('Value')  
plt.ylabel('Frequency')  

# 子图 2: 饼图  
plt.subplot(2, 2, 2)  # 2行2列的第二个子图  
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen'])  
plt.axis('equal')  # 使饼图为圆形  
plt.title('Pie Chart')  

# 子图 3: 散点图  
plt.subplot(2, 2, 3)  # 2行2列的第三个子图  
plt.scatter(x, y, c='purple', marker='o')  
plt.title('Scatter Plot')  
plt.xlabel('X')  
plt.ylabel('Y')  

# 子图 4: 条形图  
plt.subplot(2, 2, 4)  # 2行2列的第四个子图  
plt.bar(categories, values, color='teal')  
plt.title('Bar Chart')  
plt.xlabel('Category')  
plt.ylabel('Value')  

plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # 调整布局，为总标题腾出空间  
plt.show()
```

## 绘图

### 折线图

```python
import pandas as pd  
import matplotlib.pyplot as plt  
import numpy as np  

# 创建示例数据  
dates = pd.date_range(start='2025-02-04', periods=7)  
sales = np.random.randint(50, 200, size=7)  # 生成 50 到 200 之间的随机销售额  

# 创建一个 Pandas DataFrame  
df = pd.DataFrame({'Date': dates, 'Sales': sales})  
df['Date'] = pd.to_datetime(df['Date'])  

# 创建图形  
plt.figure(figsize=(10, 6), dpi=400)  

# 绘制折线图  
plt.plot(df['Date'], df['Sales'], marker='o', linestyle='-', color='skyblue', alpha=0.6)  

# 添加数据点标签  
for x, y in zip(df['Date'], df['Sales']):  
    plt.text(x, y, str(y), ha='center', va='bottom')  

# 设置标题和标签  
plt.title('Daily Sales Over 7 Days', fontsize=16)  
plt.xlabel('Date', fontsize=12)  
plt.ylabel('Sales', fontsize=12)  

# 格式化日期轴  
import matplotlib.dates as mdates  
plt.gca().xaxis.set_major_locator(mdates.DayLocator())  # 获取当前轴  
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))  # 获取当前轴  
plt.xticks(rotation=45)  

# 添加网格线  
plt.grid(True, linestyle='--', alpha=0.5)  

# 调整布局  
plt.tight_layout()  

# 显示图形  
plt.show()
```

### 饼图

```python
import matplotlib.pyplot as plt  

# 数据到饼图  
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'  
sizes = [15, 30, 45, 10]  
explode = (0, 0.1, 0, 0)  # 将第二个切片 "Hogs" 突出显示  

fig1, ax1 = plt.subplots()  
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',  
        shadow=True, startangle=90)  
ax1.axis('equal')  # 确保饼图绘制成一个圆。  

plt.show()
```

### 直方图

```python
import matplotlib.pyplot as plt  
import numpy as np  

# 示例数据：假设你有一些考试分数  
data = np.random.normal(70, 15, 200)  # 生成 200 个平均值为 70，标准差为 15 的正态分布数据  

# 创建画布  
plt.figure(figsize=(8, 6), dpi=400)  
# 画图
plt.hist(data, bins=30, color='skyblue', edgecolor='black')  # 绘制直方图  ，bins控制柱子数量
# 添加标签和标题  
plt.xlabel('Exam Score')  
plt.ylabel('Frequency')  
plt.title('Distribution of Exam Scores')  

# 添加网格线  
plt.grid(axis='y', alpha=0.75)  

# 显示图形  
plt.show()
```

```python
import matplotlib.pyplot as plt  
import numpy as np  
from matplotlib import cm  # 导入颜色映射  

# 示例数据：假设你有一些考试分数  
data = np.random.normal(70, 15, 200)  # 生成 200 个平均值为 70，标准差为 15 的正态分布数据  

# 创建直方图  
plt.figure(figsize=(10, 6), dpi=400)  # 设置图形大小和分辨率  
n, bins, patches = plt.hist(data, bins=30, edgecolor='black')  # 绘制直方图  

# 使用颜色映射设置渐变色  
cmap = cm.get_cmap('viridis')  # 选择一个颜色映射  
bin_centers = 0.5 * (bins[:-1] + bins[1:])  # 计算每个柱子的中心位置  
colornorm = plt.Normalize(vmin=min(bin_centers), vmax=max(bin_centers))  # 归一化颜色映射  
for i, patch in enumerate(patches):  
    color = cmap(colornorm(bin_centers[i]))  # 获取每个柱子的颜色  
    patch.set_facecolor(color)  # 设置柱子的颜色  

# 添加标签和标题  
plt.xlabel('Exam Score')  
plt.ylabel('Frequency')  
plt.title('Distribution of Exam Scores')  

# 添加网格线  
plt.grid(axis='y', alpha=0.75)  

# 显示图形  
plt.show()
```

### 条形图

```python
import matplotlib.pyplot as plt  
import numpy as np  

# 示例数据  
products = ['Product A', 'Product B', 'Product C', 'Product D']  
sales_dept1 = [20, 35, 30, 35]  
sales_dept2 = [25, 32, 34, 20]  

# 设置条形图的宽度  
bar_width = 0.30  

# 设置条形图的位置  
r1 = np.arange(len(products))  
r2 = [x + bar_width for x in r1]  

# 创建条形图  
plt.figure(figsize=(10, 6), dpi=400)  # 设置图形大小  
plt.bar(r1, sales_dept1, color='skyblue', width=bar_width, edgecolor='black', label='Department 1')  
plt.bar(r2, sales_dept2, color='salmon', width=bar_width, edgecolor='black', label='Department 2')  

# 添加标签和标题  
plt.xlabel('Products')  
plt.ylabel('Sales')  
plt.title('Sales Comparison by Department')  
plt.xticks([r + bar_width/2 for r in range(len(products))], products)  # 设置x轴刻度标签  
plt.legend()  

# 添加网格线  
plt.grid(axis='y', alpha=0.75)  

# 显示图形  
plt.tight_layout()  # 调整布局，避免标签重叠  
plt.show()
```

## 概率分布

### 正态分布

```python
import matplotlib.pyplot as plt  
import numpy as np  

# 设置参数  
mu = 0  # 平均值  
sigma = 1  # 标准差  
num_samples = 1000  # 样本数量  

# 生成正态分布数据  
data = np.random.normal(mu, sigma, num_samples)  

# 绘制直方图  
plt.figure(figsize=(8, 6), dpi=400)  
plt.hist(data, bins=50, density=True, alpha=0.6, color='skyblue')  

# 添加标签和标题  
plt.xlabel('Value')  
plt.ylabel('Probability Density')  
plt.title('Normal Distribution (Histogram)')  

# 添加网格线  
plt.grid(axis='y', alpha=0.75)  

# 显示图形  
plt.show()
```
